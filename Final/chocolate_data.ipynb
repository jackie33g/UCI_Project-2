```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as sts
from sqlalchemy import create_engine
```

```python
choco = "data/chocolate.csv"
choco_df = pd.read_csv(choco) 
choco_df.head()
```

```python
choco_df_clean = choco_df[['ID','company_location','review_date','country_of_bean_origin','cocoa_percent','rating']]
choco_df_clean.head()
```

```python
choco_year = choco_df_clean.loc[(choco_df_clean['review_date'] >=2015) & (choco_df_clean['review_date']<=2019)]
choco_year.head()
```

```python
choco_year.describe()
```

```python
choco_year_new= choco_year[choco_year.country_of_bean_origin !='Blend']
choco_year_new.head()
```

```python
choco_year_new.describe()
```

```python
choco_year_new1 = choco_year_new.replace('U.s.a.','U.S.A')
```

```python
print(choco_year_new1.loc[[1290]])
```

```python
choco_year_new1 = choco_year_new1.replace('U.k.','U.K.')
```

```python
print(choco_year_new1.loc[[567]])
```

```python
choco_year_new1 = choco_year_new1.replace('U.a.e.','U.A.E.')
```

```python
choco_year_new1 = choco_year_new1.replace('Sao tome','Sao tome & principe')
```

```python
print(choco_year_new1.loc[[653]])
```

```python
id_df = choco_year_new1[['ID','review_date']]
id_df.head()
```

```python
dependency_chart = choco_year_new1[['ID','company_location','country_of_bean_origin']]
dependency_chart.head()
```

```python
scatterplot_chart = choco_year_new1[['ID','company_location','country_of_bean_origin','rating']]
scatterplot_chart.head()
```

```python
donut_chart = choco_year_new1[['ID','company_location','cocoa_percent','rating']]
donut_chart.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/Chocolate_DB"
engine = create_engine(f'postgresql+psycopg2://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
id_df.to_sql(name='id_df', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from id_df', con=engine).head()
```

```python
dependency_chart.to_sql(name='dependency_chart', con=engine, if_exists='append', index=False)

```

```python
pd.read_sql_query('select * from dependency_chart', con=engine).head()
```

```python
scatterplot_chart.to_sql(name='scatterplot_chart', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from scatterplot_chart', con=engine).head()
```

```python
donut_chart.to_sql(name='donut_chart', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from donut_chart', con=engine).head()
```
